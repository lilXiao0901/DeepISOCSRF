```python
# -*- coding: utf-8 -*-

# --- 1. 匯入核心函式庫 ---
# 匯入資料處理、數值計算、圖表繪製及機器學習所需的核心函式庫。
import pandas as pd  # 用於資料處理，提供 DataFrame 結構，方便進行資料讀取、清洗與轉換。
import numpy as np   # 提供高效能的陣列與矩陣運算，是科學計算的基礎。
import matplotlib.pyplot as plt # 用於繪製靜態、互動式視覺化圖表。
import tensorflow as tf # Google 開發的深度學習框架，用於建立與訓練神經網路模型。
import time          # 用於追蹤程式執行時間 (此腳本中未直接使用，但保留以備不時之需)。
from sklearn.model_selection import train_test_split, StratifiedKFold # 用於資料集切分，StratifiedKFold 確保交叉驗證中各類別比例均衡。
from sklearn.preprocessing import StandardScaler # 用於特徵標準化，使不同尺度的特徵具有可比性，有助於模型收斂。
from sklearn.ensemble import IsolationForest   # 一種無監督的異常偵測演算法，適合用來識別離群值或異常行為。
from sklearn.metrics import confusion_matrix, classification_report, roc_curve, auc # 模型評估指標，用於衡量分類模型的性能。
from sklearn.utils.class_weight import compute_class_weight # 用於計算類別權重，解決資料不平衡問題。
from sklearn.linear_model import LogisticRegression # 匯入邏輯斯迴歸，這裡主要用於 L1 懲罰項進行特徵篩選。
from sklearn.feature_selection import SelectFromModel # 一個 meta-transformer，可根據模型的重要性權重來選擇特徵。
from sklearn.manifold import TSNE # 一種非線性降維演算法，常用於高維資料的視覺化。
from tensorflow.keras.utils import to_categorical # 將類別標籤轉換為 one-hot 編碼，以符合神經網路輸出層格式。
from tensorflow.keras.models import Sequential # Keras 的循序模型 API，用於堆疊神經網路層。
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Dropout, Dense, BatchNormalization, Input, Flatten # Keras 提供的各種神經網路層。
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau # Keras 回呼函式，用於在訓練過程中動態調整策略，如提早停止、降低學習率。
from tensorflow.keras.optimizers import Adam # 一種常用的梯度下降優化演算法。
from tensorflow.keras import backend as K # Keras 的後端引擎，用於自定義損失函數等底層操作。
import seaborn as sns # 基於 Matplotlib 的進階圖表函式庫，提供更美觀的視覺化效果。

# --- 2. 處理選用函式庫 imblearn ---
# 嘗試匯入 imbalanced-learn 函式庫中的 BorderlineSMOTE。
# 為什麼要這樣做？ imblearn 不是標準套件，使用者可能未安裝。
# 透過 try-except 結構，可以讓程式在缺少此函式庫時仍能運行（只是不執行資料平衡），並給予使用者明確的安裝提示，增強程式的穩健性。
try:
    from imblearn.over_sampling import BorderlineSMOTE
except ImportError:
    print("警告：找不到 'imblearn' 函式庫。資料將不會使用 SMOTE 進行平衡。")
    print("您可以透過執行以下指令來安裝：'pip install imbalanced-learn'")
    BorderlineSMOTE = None # 將 BorderlineSMOTE 設為 None，後續程式可透過檢查此變數判斷是否執行 SMOTE。

# --- 3. 全域圖表設定 ---
# 設定 Matplotlib 的全域參數，以確保圖表能正確顯示繁體中文。
# 為什麼要這樣做？ Matplotlib 預設字體不支援中文字元，若不設定，圖表中的中文會顯示為方塊亂碼。
plt.rcParams['font.sans-serif'] = ['Microsoft JhengHei'] # 設定預設字體為微軟正黑體。
plt.rcParams['axes.unicode_minus'] = False # 解決負號無法正常顯示的問題。

# --- 4. 視覺化輔助函式 ---
# 這些函式將繪圖邏輯封裝起來，讓主流程程式碼更簡潔、易讀，並確保圖表風格一致。

def visualize_data_distribution(X, y, title, feature_name='reqLen', stage=''):
    """
    功能：視覺化指定特徵的資料分佈與整體類別分佈。
    為什麼這樣設計？
    - 同時展示特徵分佈（直方圖）與類別平衡狀況（長條圖）有助於快速理解資料特性。
    - 特徵分佈可以觀察正常與攻擊請求在某個維度上是否有差異。
    - 類別分佈則能直觀地看出資料是否不平衡，這對後續的模型訓練策略至關重要。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    # 進行防呆設計，若指定的特徵名稱不存在，則改用第一個可用的特徵。
    if feature_name not in X.columns:
        if not X.columns.empty:
            original_feature = feature_name
            feature_name = X.columns[0]
            print(f"⚠️ 警告：預設特徵 '{original_feature}' 不存在，將改用第一個可用特徵 '{feature_name}' 進行繪圖。")
        else:
            print(f"❌ 錯誤：DataFrame 中沒有任何特徵可供繪圖。")
            return
    plt.figure(figsize=(12, 5)); plt.suptitle(title, fontsize=16)
    # 子圖一：繪製特徵的直方圖，比較正常請求與 CSRF 攻擊的差異
    plt.subplot(1, 2, 1); sns.histplot(data=X[y == 0], x=feature_name, bins=50, color='blue', alpha=0.6, label='正常請求'); sns.histplot(data=X[y == 1], x=feature_name, bins=50, color='red', alpha=0.6, label='CSRF 攻擊'); plt.title(f"'{feature_name}' 特徵分佈"); plt.xlabel(feature_name); plt.ylabel("數量"); plt.legend()
    # 子圖二：繪製類別分佈的長條圖
    plt.subplot(1, 2, 2); class_counts = pd.Series(y).value_counts().sort_index(); class_labels = ['正常請求', 'CSRF 攻擊']; bars = plt.bar(class_labels, class_counts, color=['blue', 'red']); plt.title(f"類別分佈"); plt.ylabel("樣本數")
    # 在長條圖上標示具體數量
    for bar in bars: yval = bar.get_height(); plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom', ha='center')
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

def plot_confusion_matrix(y_true, y_pred, title):
    """
    功能：繪製混淆矩陣。
    為什麼需要混淆矩陣？
    - 它能清晰地展示模型預測的四種情況：真陽性(TP)、假陽性(FP)、真陰性(TN)、假陰性(FN)。
    - 對於評估模型的錯誤類型（例如，是把攻擊誤判為正常，還是把正常誤判為攻擊）非常重要。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    cm = confusion_matrix(y_true, y_pred); plt.figure(figsize=(6, 5)); sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['正常請求', 'CSRF 攻擊'], yticklabels=['正常請求', 'CSRF 攻擊'], annot_kws={"size": 14}); plt.title(title, fontsize=15); plt.xlabel('預測標籤', fontsize=12); plt.ylabel('真實標籤', fontsize=12); plt.show()

def plot_metrics_report(report_dict, title):
    """
    功能：將 classification_report 的結果視覺化。
    為什麼要視覺化報告？
    - 數字報告雖然詳盡，但圖表能更直觀地比較不同類別在精確率(Precision)、召回率(Recall)、F1 分數(F1-score)上的表現。
    - 尤其在比較多個模型或多折交叉驗證結果時，圖表更具可讀性。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    # 將 classification_report 的字典格式轉換為適合繪圖的 DataFrame 格式。
    classes_map = {'Normal': '正常請求', 'CSRF': 'CSRF 攻擊'}; metrics_map = {'precision': '精確率', 'recall': '召回率', 'f1-score': 'F1 分數'}; classes = [c for c in report_dict.keys() if c in ['Normal', 'CSRF']]; metrics_to_plot = ['precision', 'recall', 'f1-score']; plot_data = { 'class': [], 'metric': [], 'value': [] }
    for cls_en in classes:
        cls_zh = classes_map.get(cls_en, cls_en)
        for metric_en in metrics_to_plot: metric_zh = metrics_map.get(metric_en, metric_en); plot_data['class'].append(cls_zh); plot_data['metric'].append(metric_zh); plot_data['value'].append(report_dict[cls_en][metric_en])
    df = pd.DataFrame(plot_data); plt.figure(figsize=(10, 6)); sns.barplot(data=df, x='metric', y='value', hue='class', palette={'正常請求': 'blue', 'CSRF 攻擊': 'red'}); plt.title(title, fontsize=15); plt.ylabel('分數', fontsize=12); plt.xlabel('評估指標', fontsize=12); plt.ylim(0, 1.05); plt.legend(title='類別'); plt.grid(axis='y', linestyle='--', alpha=0.7); plt.show()

def plot_merged_isolation_forest_scores(all_scores, all_labels, title):
    """
    功能：繪製孤立森林產生的異常分數分佈圖。
    為什麼要看這個分佈？
    - 理想情況下，正常請求（藍色）和 CSRF 攻擊（紅色）的異常分數應該有明顯的分界。
    - 這個圖可以幫助我們判斷孤立森林作為特徵提取器的有效性。如果兩者分佈重疊嚴重，代表這個特徵的區分度不高。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    plt.figure(figsize=(10, 6))
    df_scores = pd.DataFrame({'scores': all_scores, 'label': all_labels})
    sns.histplot(data=df_scores[df_scores['label'] == 0], x='scores', color='blue', alpha=0.6, label='正常請求', bins=50)
    sns.histplot(data=df_scores[df_scores['label'] == 1], x='scores', color='red', alpha=0.6, label='CSRF 攻擊', bins=50)
    plt.title(title, fontsize=15)
    # --- 更新圖表標籤 ---
    # 標籤明確指出分數越高代表越異常，幫助理解圖表的意義。
    plt.xlabel('異常分數 (越高越異常)', fontsize=12)
    # --- 修改結束 ---
    plt.ylabel('數量', fontsize=12); plt.legend(); plt.show()

def plot_merged_training_history(histories, title):
    """
    功能：將交叉驗證中每一次的訓練歷史（損失、準確率）合併，並繪製平均值與標準差。
    為什麼要合併繪製？
    - 單次的訓練曲線可能存在隨機性。
    - 透過平均多折的結果，可以得到一個更穩定、更具代表性的學習曲線，反映模型整體的學習趨勢。
    - 標準差（陰影區域）則顯示了模型在不同資料子集上的性能穩定性。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    plt.figure(figsize=(12, 5)); plt.suptitle(title, fontsize=16)
    metrics = ['loss', 'val_loss', 'accuracy', 'val_accuracy']
    plot_info = {'loss': ('模型損失', '訓練損失', 'blue'), 'val_loss': ('模型損失', '驗證損失', 'orange'), 'accuracy': ('模型準確率', '訓練準確率', 'blue'), 'val_accuracy': ('模型準確率', '驗證準確率', 'orange')}
    subplots = {'loss': (1, 2, 1), 'accuracy': (1, 2, 2)}
    # 找到所有折中最小的訓練週期數，以統一繪圖的 X 軸。
    min_epochs = min(len(h.history['loss']) for h in histories)
    for metric_key, subplot_pos in subplots.items():
        plt.subplot(*subplot_pos)
        # 繪製訓練指標和驗證指標
        for metric in [m for m in metrics if m.replace('val_', '') == metric_key]:
            title_p, label, color = plot_info[metric]
            # 取出所有折的該指標數據
            all_folds_metric = [h.history[metric][:min_epochs] for h in histories]
            # 計算平均值和標準差
            mean_metric = np.mean(all_folds_metric, axis=0); std_metric = np.std(all_folds_metric, axis=0)
            epochs_range = range(1, min_epochs + 1)
            # 繪製平均曲線
            plt.plot(epochs_range, mean_metric, label=f'平均{label}', color=color)
            # 繪製標準差區間
            plt.fill_between(epochs_range, mean_metric - std_metric, mean_metric + std_metric, color=color, alpha=0.2, label=f'{label} 標準差')
        plt.title(title_p); plt.xlabel("週期 (Epoch)"); plt.ylabel(title_p.split(' ')[-1]); plt.legend(); plt.grid(True, linestyle='--', alpha=0.6)
    plt.tight_layout(rect=[0, 0.03, 1, 0.95]); plt.show()

def plot_roc_curve(y_true, y_pred_proba, title):
    """
    功能：繪製 ROC 曲線並計算 AUC 值。
    為什麼 ROC/AUC 很重要？
    - ROC 曲線展示了在不同分類閾值下，真陽性率 (TPR) 與偽陽性率 (FPR) 之間的權衡關係。
    - AUC (曲線下面積) 是一個單一數值，衡量模型整體區分正負樣本的能力，不受特定閾值影響。AUC 越接近 1，模型性能越好。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    fpr, tpr, _ = roc_curve(y_true, y_pred_proba); roc_auc = auc(fpr, tpr)
    plt.figure(figsize=(8, 6)); plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'匯總 ROC 曲線 (AUC = {roc_auc:.4f})'); plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--'); plt.xlim([0.0, 1.0]); plt.ylim([0.0, 1.05]); plt.xlabel("偽陽性率 (False Positive Rate)", fontsize=12); plt.ylabel("真陽性率 (True Positive Rate)", fontsize=12); plt.title(title, fontsize=15); plt.legend(loc="lower right"); plt.grid(True); plt.show()

def plot_feature_importance(importances, feature_names, title):
    """
    功能：視覺化特徵重要性。
    為什麼要了解特徵重要性？
    - 它可以幫助我們理解模型是根據哪些特徵來做決策的，提升模型的可解釋性。
    - 也可以用來驗證我們的特徵選擇是否合理，或者發現一些意想不到的關鍵特徵。
    """
    print(f"\n--- 正在產生圖表: {title} ---")
    df_importances = pd.DataFrame({'feature': feature_names, 'importance': importances})
    df_importances = df_importances.sort_values(by='importance', ascending=False)
    plt.figure(figsize=(10, 8)); sns.barplot(x='importance', y='feature', data=df_importances, palette='viridis'); plt.title(title, fontsize=15); plt.xlabel('重要性分數 (L1 係數絕對值)', fontsize=12); plt.ylabel('特徵', fontsize=12); plt.grid(axis='x', linestyle='--', alpha=0.6); plt.tight_layout(); plt.show()

def plot_tsne_visualization(X_high_dim, y_true, anomaly_scores, title_prefix):
    """
    功能：使用 t-SNE 將高維特徵降至二維並進行視覺化。
    為什麼使用 t-SNE？
    - 人類無法直接觀察高維空間。t-SNE 是一種強大的降維工具，能將高維空間中的樣本點映射到二維平面上，同時盡量保持它們的局部結構。
    - 透過 t-SNE 圖，我們可以直觀地看到：
      1. 依真實標籤著色：正常樣本和攻擊樣本是否在特徵空間中形成了可區分的群集。
      2. 依孤立森林分數著色：異常分數高的點是否與真實的攻擊樣本群集重合，再次驗證孤立森林特徵的有效性。
    """
    print(f"\n--- 正在產生 t-SNE 降維與視覺化 (此過程可能需要一些時間) ---")
    # n_components=2 表示降到二維。perplexity 是一個重要參數，與每個點的近鄰數量有關。
    tsne = TSNE(n_components=2, random_state=42, perplexity=min(30, len(X_high_dim)-1))
    X_tsne = tsne.fit_transform(X_high_dim)
    # 將降維結果與標籤、異常分數合併成一個 DataFrame，方便繪圖。
    df_tsne = pd.DataFrame(X_tsne, columns=['t-SNE 1', 't-SNE 2'])
    df_tsne['label'] = y_true; df_tsne['label_str'] = df_tsne['label'].map({0: '正常請求', 1: 'CSRF 攻擊'}); df_tsne['anomaly_score'] = anomaly_scores
    plt.figure(figsize=(16, 7))
    # 子圖一：根據真實標籤著色
    ax1 = plt.subplot(1, 2, 1); sns.scatterplot(x='t-SNE 1', y='t-SNE 2', hue='label_str', data=df_tsne, palette={'正常請求': 'blue', 'CSRF 攻擊': 'red'}, alpha=0.7, ax=ax1); ax1.set_title(f'{title_prefix} (依真實標籤著色)', fontsize=15); ax1.legend(title='類別'); ax1.grid(True, linestyle='--', alpha=0.5)
    # 子圖二：根據孤立森林分數著色
    ax2 = plt.subplot(1, 2, 2); scatter = ax2.scatter(df_tsne['t-SNE 1'], df_tsne['t-SNE 2'], c=df_tsne['anomaly_score'], cmap='viridis_r', alpha=0.7) # 使用 viridis_r 反轉色譜，讓顏色更深的代表分數更高（更異常）。
    # --- 更新圖表標籤 ---
    plt.colorbar(scatter, label='異常分數 (越高越異常)')
    # --- 修改結束 ---
    ax2.set_title(f'{title_prefix} (依孤立森林分數著色)', fontsize=15); ax2.grid(True, linestyle='--', alpha=0.5); plt.tight_layout(); plt.show()

# --- 5. 核心訓練評估函式 (用於特徵搜尋) ---
def train_and_evaluate_for_search(X, y, with_augmentation_smote=True):
    """
    功能：這是一個簡化版的訓練與評估流程，專門用於快速測試不同特徵組合的性能。
    為什麼要建立這個函式？
    - 特徵選擇階段需要對多種特徵組合進行評估，如果每次都執行完整的交叉驗證會非常耗時。
    - 這個函式使用一次性的 train_test_split 來快速得到一個 F1 分數，作為評估該特徵組合好壞的依據，大大加快了特徵選擇的速度。
    """
    # 與主流程類似，進行資料增強與平衡
    if with_augmentation_smote and BorderlineSMOTE is not None:
        def augment_data(X_df, y_s, noise_level=0.1, class_label=0):
            # 對多數類（正常請求）添加微小雜訊來生成新樣本，這是一種簡單的資料增強。
            mask = y_s == class_label; X_class = X_df[mask].copy()
            noise = np.random.normal(0, noise_level, X_class.shape) * X_class.std(axis=0).values
            X_aug = X_class + noise
            return pd.concat([X_df, X_aug], ignore_index=True), pd.concat([y_s, y_s[mask]], ignore_index=True)
        X_aug, y_aug = augment_data(X, y)
        smote = BorderlineSMOTE(random_state=42, k_neighbors=5); X_resampled_np, y_resampled = smote.fit_resample(X_aug, y_aug)
        X_final_df = pd.DataFrame(X_resampled_np, columns=X.columns); y_final = y_resampled
    else: X_final_df, y_final = X, y

    # 使用 stratify 參數確保切分後的訓練集和測試集有著相同的類別比例。
    X_train_df, X_test_df, y_train, y_test = train_test_split(X_final_df, y_final, test_size=0.2, random_state=42, stratify=y_final)
    # 特徵標準化
    scaler = StandardScaler(); X_train_scaled = scaler.fit_transform(X_train_df); X_test_scaled = scaler.transform(X_test_df)
    # 訓練孤立森林並提取異常分數作為新特徵
    iso_forest = IsolationForest(contamination='auto', random_state=42); iso_forest.fit(X_train_scaled)
    # --- 修改開始: 分數乘以-1 ---
    # 為什麼要乘以 -1？ scikit-learn 的 decision_function 預設是「分數越低越異常」。
    # 乘以 -1 後，變為「分數越高越異常」，這更符合一般直覺，方便後續的理解與視覺化。
    train_anomaly_scores = -iso_forest.decision_function(X_train_scaled).reshape(-1, 1)
    test_anomaly_scores = -iso_forest.decision_function(X_test_scaled).reshape(-1, 1)
    # --- 修改結束 ---
    # 將異常分數與原特徵合併，形成最終的特徵集
    X_train_final = np.hstack((X_train_scaled, train_anomaly_scores)); X_test_final = np.hstack((X_test_scaled, test_anomaly_scores))
    # 為 CNN 準備 3D 輸入格式 (samples, timesteps, features)
    X_train_cnn = np.reshape(X_train_final, (X_train_final.shape[0], X_train_final.shape[1], 1)); X_test_cnn = np.reshape(X_test_final, (X_test_final.shape[0], X_test_final.shape[1], 1))
    y_train_cat = to_categorical(y_train)
    # 計算類別權重，在模型訓練時賦予少數類別更高的權重，以緩解資料不平衡的影響。
    class_weight_dict = dict(enumerate(compute_class_weight('balanced', classes=np.unique(y_train.values), y=y_train.values)))

    # --- 自定義 Focal Loss 損失函數 ---
    def focal_loss(gamma=2.0, alpha=0.25):
        """
        為什麼使用 Focal Loss？
        - 它是標準交叉熵損失的改良版，專為解決類別不平衡問題設計。
        - 它透過一個調節因子 (1-p_t)^gamma，降低了大量易於分類樣本（如大部分正常請求）對損失的貢獻。
        - 這使得模型能夠更專注於學習那些難以區分的、錯誤分類的樣本（通常是少數類的攻擊樣本），從而提升模型的召回率和整體性能。
        """
        def focal_loss_fixed(y_true, y_pred):
            y_true_float = tf.cast(y_true, tf.float32); epsilon = K.epsilon(); y_pred = K.clip(y_pred, epsilon, 1. - epsilon)
            cross_entropy = -y_true_float * K.log(y_pred); loss = alpha * K.pow(1 - y_pred, gamma) * cross_entropy
            return K.mean(loss, axis=-1)
        return focal_loss_fixed

    # 建立 CNN 模型
    model = Sequential([Input(shape=(X_train_cnn.shape[1], 1)), Conv1D(64, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), MaxPooling1D(pool_size=2), Dropout(0.5), Conv1D(128, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), Dropout(0.5), Conv1D(256, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), Dropout(0.5), Flatten(), Dense(128, activation='relu'), Dropout(0.5), Dense(2, activation='softmax')])
    model.compile(optimizer=Adam(learning_rate=1e-3), loss=focal_loss(), metrics=['accuracy'])
    # 設定回呼函式
    early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0)
    reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=0)
    # 訓練模型
    model.fit(X_train_cnn, y_train_cat, epochs=100, batch_size=16, validation_split=0.1, class_weight=class_weight_dict, callbacks=[early_stopping, reduce_lr], verbose=0)
    # 預測並評估
    y_pred = np.argmax(model.predict(X_test_cnn), axis=1)
    report_dict = classification_report(y_test, y_pred, output_dict=True, zero_division=0)
    # 回傳 F1 分數作為此特徵組合的性能指標
    return report_dict['weighted avg']['f1-score']

# --- 6. 主流程函式 ---
def run_final_pipeline_v16(with_augmentation_smote=True):
    """
    這是整個 CSRF 偵測流程的主函式，它整合了資料前處理、特徵工程、模型訓練與評估的所有步驟。
    """
    # --- 資料讀取與前處理 ---
    try:
        data = pd.read_csv('mitch.csv')
    except FileNotFoundError:
        print("❌ 錯誤：找不到 'mitch.csv' 檔案。"); return
    
    # 為什麼要移除極端值？
    # 極端值（或稱離群值）可能會對 StandardScaler 和模型訓練產生不成比例的巨大影響，導致模型性能下降。
    # 使用 99% 分位數是一種常見且穩健的作法，可以濾除最極端的 1% 數據，同時保留大部分資料分佈。
    q99_reqlen = data['reqLen'].quantile(0.99); q99_params = data['numOfParams'].quantile(0.99)
    data_cleaned = data[(data['reqLen'] <= q99_reqlen) & (data['numOfParams'] <= q99_params)].dropna()

    # 選擇 Mitch 論文中提到的前 11 個重要特徵
    mitch_top11_features = ['reqLen', 'numOfParams', 'isGET', 'isPOST', 'tokenInParams', 'numOfBools', 'logoutInPath', 'numOfBlobs', 'passwordInParams', 'numOfIds', 'deleteInPath']
    missing_features = [f for f in mitch_top11_features if f not in data_cleaned.columns]
    if missing_features: print(f"❌ 錯誤：CSV 檔案中缺少以下必要特徵：{missing_features}"); return
    
    # 準備特徵 (X) 與標籤 (y)
    X_candidate = data_cleaned[mitch_top11_features]; y_candidate = data_cleaned['flag']
    # 過濾掉標籤不是 'y' 或 'n' 的無效資料
    valid_indices = y_candidate.isin(['y', 'n'])
    # 將 'y'/'n' 映射為 1/0，因為模型需要數值輸入
    X_candidate = X_candidate[valid_indices].reset_index(drop=True); y_candidate = y_candidate[valid_indices].map({'y': 1, 'n': 0}).reset_index(drop=True)
    # 視覺化原始資料分佈
    visualize_data_distribution(X_candidate, y_candidate, "1. 原始 Mitch 資料集分佈圖")

    # --- 階段一：自動化特徵篩選 ---
    # 為什麼要進行特徵篩選？
    # - 並非所有特徵都對模型有益，無關或冗餘的特徵可能會引入噪聲，導致模型過擬合。
    # - 篩選出最關鍵的特徵可以簡化模型，加快訓練速度，並可能提升性能。
    # 這裡的策略是：系統性地測試使用 5 到 11 個特徵時模型的性能，找出最佳的特徵數量。
    print("\n" + "="*60 + "\n🚀 階段一：開始系統化特徵篩選與重要性分析...\n" + "="*60)
    performance_results = []; feature_counts_to_test = range(5, 12)
    
    # 為什麼使用帶有 L1 懲罰項的邏輯斯迴歸？
    # - L1 懲罰項（Lasso）在訓練過程中會將不重要特徵的係數壓縮至零，因此它本身就是一種強大的特徵選擇工具。
    # - 相較於隨機森林等方法，它計算速度快，非常適合用於這個自動化篩選的循環中。
    lr_selector = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, class_weight='balanced')
    
    for k in feature_counts_to_test:
        print(f"\n--- 正在測試篩選 {k} 個特徵時的性能 ---")
        # SelectFromModel 會根據 lr_selector 訓練後得到的係數大小，選出最重要的 k 個特徵。
        selector = SelectFromModel(lr_selector, max_features=k, prefit=False).fit(X_candidate, y_candidate)
        selected_features = X_candidate.columns[selector.get_support()].tolist()
        X_selected = X_candidate[selected_features]
        # 使用簡化版的評估函式快速得到 F1 分數。
        f1 = train_and_evaluate_for_search(X_selected, y_candidate, with_augmentation_smote)
        print(f"👉 使用 {k} 個特徵的 F1 分數為: {f1:.4f}")
        performance_results.append({'k': k, 'f1_score': f1})
    
    # 視覺化不同特徵數量下的性能變化
    results_df = pd.DataFrame(performance_results)
    plt.figure(figsize=(10, 6)); plt.plot(results_df['k'], results_df['f1_score'], marker='o', linestyle='--'); plt.title('不同特徵數量下的模型性能 (F1 分數)'); plt.xlabel('選擇的特徵數量 (k)'); plt.ylabel('F1 分數'); plt.xticks(feature_counts_to_test); plt.grid(True); plt.show()
    
    # 找出 F1 分數最高時對應的特徵數量
    best_k_row = results_df.loc[results_df['f1_score'].idxmax()]
    best_k = int(best_k_row['k'])
    print(f"\n✅ 階段一完成！找到的最佳特徵數量為: {best_k} 個。")

    # 使用全部 11 個特徵訓練一次 L1 邏輯斯迴歸，以視覺化所有特徵的重要性排序。
    lr_importance = LogisticRegression(penalty='l1', solver='liblinear', random_state=42, class_weight='balanced')
    lr_importance.fit(X_candidate, y_candidate)
    importances = np.abs(lr_importance.coef_[0])
    plot_feature_importance(importances, mitch_top11_features, "2. Mitch 11 特徵重要程度圖")

    # --- 階段二：準備交叉驗證資料 ---
    # 為什麼要進行資料增強與平衡？
    # - 原始資料中攻擊樣本（正類）遠少於正常樣本（負類），模型容易偏向多數類，導致對攻擊的檢測率（召回率）很低。
    # - 這裡採用了兩階段策略：
    #   1. 資料增強 (Data Augmentation): 對多數類（正常請求）加入微小雜訊，產生更多樣化的正常樣本，幫助模型學習更穩健的決策邊界。
    #   2. Borderline-SMOTE: 一種先進的過採樣技術，它專注於在類別邊界附近生成新的少數類樣本。這比隨機生成更有效，因為這些邊界樣本是模型最容易混淆的，強化對它們的學習有助於提升分類性能。
    print("\n" + "="*60 + f"\n🚀 階段二：使用 {best_k} 個特徵準備交叉驗證資料...\n" + "="*60)
    # 使用階段一找到的最佳 k 值，正式選出最終要使用的特徵。
    selector = SelectFromModel(lr_selector, max_features=best_k, prefit=False).fit(X_candidate, y_candidate)
    final_selected_features = X_candidate.columns[selector.get_support()].tolist()
    print(f"最終用於交叉驗證的 {best_k} 個特徵為: {final_selected_features}")
    X_final_selected = X_candidate[final_selected_features]

    # 定義資料增強函式
    def augment_data(X_df, y_s, noise_level=0.1, class_label=0):
        mask = y_s == class_label; X_class = X_df[mask].copy()
        # 雜訊的大小是根據每個特徵的標準差來縮放的，確保加入的雜訊與特徵本身的尺度相稱。
        noise = np.random.normal(0, noise_level, X_class.shape) * X_class.std(axis=0).values
        X_aug = X_class + noise
        return pd.concat([X_df, X_aug], ignore_index=True), pd.concat([y_s, y_s[mask]], ignore_index=True)

    if with_augmentation_smote and BorderlineSMOTE is not None:
        X_aug, y_aug = augment_data(X_final_selected, y_candidate)
        visualize_data_distribution(X_aug, y_aug, "3. 資料增廣後資料分佈圖 (加入雜訊)")
        smote = BorderlineSMOTE(random_state=42, k_neighbors=5)
        X_resampled_np, y_resampled = smote.fit_resample(X_aug, y_aug)
        X_for_cv = pd.DataFrame(X_resampled_np, columns=final_selected_features)
        y_for_cv = y_resampled
        visualize_data_distribution(X_for_cv, y_for_cv, "4. Borderline-SMOTE 後資料分佈圖")
    else:
        print("警告：未執行資料增強與 SMOTE。"); X_for_cv, y_for_cv = X_final_selected, y_candidate

    # --- 階段三：執行 10 折交叉驗證 ---
    # 為什麼要用交叉驗證？
    # - 只做一次 train_test_split 的結果可能帶有偶然性，受到資料劃分方式的影響。
    # - 10 折交叉驗證將資料分成 10 份，輪流將其中 9 份作為訓練集、1 份作為測試集，進行 10 次獨立的訓練與評估。
    # - 最後將 10 次的結果匯總，可以得到一個更穩定、更可靠的模型性能評估，減少隨機性帶來的偏差。
    # - 使用 StratifiedKFold 確保每一折中的類別比例都與原始資料集相似，這在不平衡資料集上至關重要。
    print("\n" + "="*60 + f"\n🚀 階段三：開始執行 10 折交叉驗證...\n" + "="*60)
    skf = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)
    # 初始化用於儲存所有折結果的列表
    all_fold_histories, all_iso_scores, all_iso_true_labels, all_iso_pred_labels = [], np.array([]), np.array([]), np.array([])
    all_cnn_true_labels, all_cnn_pred_labels, all_cnn_pred_probas = np.array([]), np.array([]), np.array([])
    all_X_test_scaled_list = []
    X_cv_values, y_cv_values = X_for_cv.values, y_for_cv.values

    # 再次定義 Focal Loss，確保在交叉驗證的循環中可用
    def focal_loss(gamma=2.0, alpha=0.25):
        def focal_loss_fixed(y_true, y_pred):
            y_true_float = tf.cast(y_true, tf.float32); epsilon = K.epsilon(); y_pred = K.clip(y_pred, epsilon, 1. - epsilon)
            cross_entropy = -y_true_float * K.log(y_pred); loss = alpha * K.pow(1 - y_pred, gamma) * cross_entropy
            return K.mean(loss, axis=-1)
        return focal_loss_fixed

    for fold, (train_idx, test_idx) in enumerate(skf.split(X_cv_values, y_cv_values)):
        print(f"--- 正在處理第 {fold+1}/10 折 ---")
        X_train, X_test = X_cv_values[train_idx], X_cv_values[test_idx]; y_train, y_test = y_cv_values[train_idx], y_cv_values[test_idx]
        
        # 特徵標準化：Scaler 必須在訓練集上 fit，然後同時 transform 訓練集和測試集。
        # 為什麼？ 這是為了防止測試集的資訊洩漏到訓練過程中。模型的任何預處理步驟都應只基於訓練資料來學習。
        scaler = StandardScaler(); X_train_scaled = scaler.fit_transform(X_train); X_test_scaled = scaler.transform(X_test)
        all_X_test_scaled_list.append(X_test_scaled)
        
        # --- 混合模型第一層：孤立森林 ---
        # 為什麼要用孤立森林作為特徵工程的一環？
        # - 這是一個「模型堆疊」(Stacking) 的思想。孤立森林作為一個無監督的異常檢測器，可以從一個新的角度來審視數據。
        # - 它產生的「異常分數」是一個高度濃縮的資訊，代表了“這個樣本有多麼不尋常”。
        # - 將這個分數作為一個新的特徵加入到原始特徵集中，可以為下游的 CNN 模型提供非常有價值的額外資訊，幫助它更好地區分正常與異常。
        iso_forest = IsolationForest(contamination='auto', random_state=42); iso_forest.fit(X_train_scaled)
        # --- 修改開始: 分數乘以-1 ---
        # 同樣，將分數轉換為「越高越異常」的直觀形式
        iso_scores_test = -iso_forest.decision_function(X_test_scaled)
        train_anomaly_scores = -iso_forest.decision_function(X_train_scaled).reshape(-1, 1)
        test_anomaly_scores = -iso_forest.decision_function(X_test_scaled).reshape(-1, 1)
        # --- 修改結束 ---
        # 孤立森林本身的預測結果 (1 表示正常, -1 表示異常)，轉換為 0/1 格式
        iso_pred_test = iso_forest.predict(X_test_scaled); iso_pred_mapped_test = np.where(iso_pred_test == 1, 0, 1)
        # 收集每一折孤立森林的結果，用於後續的整體評估
        all_iso_scores = np.concatenate([all_iso_scores, iso_scores_test]); all_iso_true_labels = np.concatenate([all_iso_true_labels, y_test]); all_iso_pred_labels = np.concatenate([all_iso_pred_labels, iso_pred_mapped_test])
        
        # 合併原始特徵和異常分數特徵
        X_train_final = np.hstack((X_train_scaled, train_anomaly_scores)); X_test_final = np.hstack((X_test_scaled, test_anomaly_scores))
        
        # --- 混合模型第二層：1D CNN ---
        # 為什麼選擇 1D CNN？
        # - 傳統上 CNN 用於處理圖像 (2D) 或時間序列 (1D)。
        # - 在這裡，我們將每個樣本的特徵向量（包含原始特徵和異常分數）視為一個一維序列。
        # - 1D CNN 的卷積核可以滑過這個序列，學習特徵之間的局部依賴關係和組合模式，這比傳統的全連接層（Dense）更能捕捉複雜的非線性關係。
        X_train_cnn = np.reshape(X_train_final, (X_train_final.shape[0], X_train_final.shape[1], 1)); X_test_cnn = np.reshape(X_test_final, (X_test_final.shape[0], X_test_final.shape[1], 1))
        y_train_cat = to_categorical(y_train)
        class_weight_dict = dict(enumerate(compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)))
        
        # 建立、編譯、訓練 CNN 模型
        model = Sequential([Input(shape=(X_train_cnn.shape[1], 1)), Conv1D(64, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), MaxPooling1D(pool_size=2), Dropout(0.5), Conv1D(128, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), Dropout(0.5), Conv1D(256, kernel_size=3, activation='relu', padding='same'), BatchNormalization(), Dropout(0.5), Flatten(), Dense(128, activation='relu'), Dropout(0.5), Dense(2, activation='softmax')])
        model.compile(optimizer=Adam(learning_rate=1e-3), loss=focal_loss(), metrics=['accuracy'])
        
        # 為什麼要用 EarlyStopping 和 ReduceLROnPlateau？
        # - EarlyStopping: 監控驗證集上的損失，如果連續多個週期（patience=10）都沒有改善，就提前終止訓練，並恢復到最佳權重時的狀態。這可以有效防止過擬合，並節省訓練時間。
        # - ReduceLROnPlateau: 監控驗證集上的損失，如果連續多個週期（patience=5）沒有改善，就自動降低學習率。這有助於模型在接近最優點時進行更精細的搜索，可能找到更好的解。
        early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True, verbose=0); reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6, verbose=0)
        history = model.fit(X_train_cnn, y_train_cat, epochs=100, batch_size=16, validation_split=0.1, class_weight=class_weight_dict, callbacks=[early_stopping, reduce_lr], verbose=0)
        
        # 收集每一折 CNN 的結果
        all_fold_histories.append(history)
        y_pred_proba_fold = model.predict(X_test_cnn) # 獲取屬於類別 1 的機率，用於繪製 ROC 曲線
        y_pred_classes_fold = np.argmax(y_pred_proba_fold, axis=1) # 獲取最終預測類別
        all_cnn_true_labels = np.concatenate([all_cnn_true_labels, y_test]); all_cnn_pred_labels = np.concatenate([all_cnn_pred_labels, y_pred_classes_fold]); all_cnn_pred_probas = np.concatenate([all_cnn_pred_probas, y_pred_proba_fold[:, 1]])
    
    print("\n✅ 階段三完成！模型訓練與評估結束。")

    # --- 階段四：匯總報告與視覺化 ---
    # 在這個階段，我們將交叉驗證中收集到的所有結果進行匯總，並產生一系列的視覺化圖表，以全面評估模型的最終性能。
    print("\n" + "="*60 + f"\n🚀 階段四：產生最終匯總報告與視覺化圖表...\n" + "="*60)
    # 合併所有測試集的 scaled data，用於 t-SNE 視覺化
    all_X_test_scaled = np.vstack(all_X_test_scaled_list)
    
    # 產生一系列評估圖表
    plot_tsne_visualization(all_X_test_scaled, all_iso_true_labels, all_iso_scores, "5. 孤立森林資料點 t-SNE 圖")
    plot_merged_isolation_forest_scores(all_iso_scores, all_iso_true_labels, "6. 合併的孤立森林異常分數分佈圖")
    plot_confusion_matrix(all_iso_true_labels, all_iso_pred_labels, "7. 合併的孤立森林混淆矩陣 (10 折匯總)")
    iso_report_dict = classification_report(all_iso_true_labels, all_iso_pred_labels, target_names=['Normal', 'CSRF'], output_dict=True, zero_division=0)
    plot_metrics_report(iso_report_dict, "8. 合併的孤立森林性能指標圖 (10 折匯總)")
    
    plot_merged_training_history(all_fold_histories, "9. 合併的模型訓練歷史圖 (10 折平均)")
    plot_confusion_matrix(all_cnn_true_labels, all_cnn_pred_labels, "10. 合併的 CNN 混淆矩陣 (10 折匯總)")
    final_report_dict = classification_report(all_cnn_true_labels, all_cnn_pred_labels, target_names=['Normal', 'CSRF'], output_dict=True, zero_division=0)
    plot_metrics_report(final_report_dict, "11. 合併的 CNN 性能指標圖 (10 折匯總)")
    plot_roc_curve(all_cnn_true_labels, all_cnn_pred_probas, "12. 合併的 CNN ROC 曲線 (10 折匯總)")

    print("\n" + "="*60 + "\n✅ 所有流程已完成！\n" + "="*60)

# --- 7. 程式進入點 ---
if __name__ == '__main__':
    # 為什麼要有這段檢查與生成假檔案的程式碼？
    # - 這是為了讓使用者在沒有 `mitch.csv` 資料集的情況下，也能夠成功運行此腳本並看到完整的流程。
    # - 它會生成一個結構相同、數據分佈相似的假檔案，確保所有後續步驟都能正常執行。
    # - 這極大地提高了程式碼的可移植性和易用性。
    try:
        pd.read_csv('mitch.csv')
    except FileNotFoundError:
        print("偵測到 'mitch.csv' 不存在，正在生成一個用於演示的假檔案...")
        num_samples = 500
        data = {'reqLen': np.random.randint(50, 500, num_samples), 'numOfParams': np.random.randint(1, 10, num_samples), 'isGET': np.random.choice([0, 1], num_samples, p=[0.3, 0.7]), 'isPOST': np.random.choice([0, 1], num_samples, p=[0.7, 0.3]), 'tokenInParams': np.random.choice([0, 1], num_samples, p=[0.8, 0.2]), 'numOfBools': np.random.randint(0, 5, num_samples), 'logoutInPath': np.random.choice([0, 1], num_samples, p=[0.95, 0.05]), 'numOfBlobs': np.random.randint(0, 3, num_samples), 'passwordInParams': np.random.choice([0, 1], num_samples, p=[0.9, 0.1]), 'numOfIds': np.random.randint(0, 4, num_samples), 'deleteInPath': np.random.choice([0, 1], num_samples, p=[0.98, 0.02]), 'flag': np.random.choice(['n', 'y'], num_samples, p=[0.85, 0.15])}
        # 為了讓假數據更逼真，手動調整一些攻擊樣本的特徵值
        attack_indices = np.where(data['flag'] == 'y')[0]
        data['reqLen'][attack_indices] = np.random.randint(200, 600, len(attack_indices)); data['numOfParams'][attack_indices] = np.random.randint(5, 15, len(attack_indices)); data['tokenInParams'][attack_indices] = 0
        df_dummy = pd.DataFrame(data); df_dummy.to_csv('mitch.csv', index=False)
        print("'mitch.csv' 假檔案已生成。")

    # 執行主流程
    run_final_pipeline_v16(with_augmentation_smote=True)
```
